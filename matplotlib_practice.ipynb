```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y_axis to a list of floats as the total fare in the US dollars accumlated for each month.
y_axis = [10.02, 23.34, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot().
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = 'Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set the y limit between 0 and 45.
plt.ylim(0, 45)

# create title
plt.title("PyBer Fare by Month")

# Add grid.
plt.grid()

# Add legend.
plt.legend()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label = 'Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set the y limit between 0 and 45.
plt.ylim(0, 45)

# create title
plt.title("PyBer Fare by Month")

# Add grid.
plt.grid()

# Add legend.
plt.legend()
```

```python
# create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.bar(x_axis, y_axis, color="green", label = 'Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# create title
plt.title("PyBer Fare by Month")

# Add legend.
plt.legend()
```

```python
# create the plot
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot and add a label for the legend.
plt.barh(x_axis, y_axis, color="magenta", label = 'Boston')
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# create title
plt.title("PyBer Fare by Month")

# Add legend.
plt.legend()
```

```python
# Create the plot with ax.plt().
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt().
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="Cyan", label='New York')
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create graph
plt.scatter(y_axis, x_axis, c='red', label= "Chicago")
# create x and y labels
plt.xlabel("Date")
plt.ylabel("Fare($)")
# create title
plt.title("Pyber Fare by Month")
# Set x limit
plt.xlim(0, 50)
# invert y_axis
plt.gca().invert_yaxis()
# Add legend
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i *3 for i in y_axis])
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
        explode=explode_values, 
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')

plt.show()
```
